```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
```

```python
avg_math = df["math score"].mean()
avg_math
```

```python
gender_avg = df.groupby("gender")["math score"].mean()

plt.figure()
plt.bar(gender_avg.index, gender_avg.values)
plt.xlabel("Gender")
plt.ylabel("Average Math Score")
plt.title("Average Math Score by Gender")
plt.show()
```

```python
plt.figure()
plt.scatter(df["math score"], df["reading score"])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math vs Reading Scores")
plt.show()
```

```python
corr = df[["math score", "reading score", "writing score"]].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Score Correlation Heatmap")
plt.show()
```

# Insights & Observations

- The average math score indicates moderate performance.
- Female students slightly outperform male students in math.
- Strong correlation exists between reading and writing scores.
- Math performance shows a positive but weaker correlation with language subjects.
